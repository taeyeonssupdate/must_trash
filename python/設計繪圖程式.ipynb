```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np
```

```python
data = [-1, -4.3, 15, 21, 31]
plt.plot(data,"b--o")  # x軸是 0,1,2,3,4
plt.show()
```

```python
x = np.linspace(0, 10, 1000)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, x, cosinus)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
plt.scatter(x, y)
plt.show()
```

```python
labels = ["Python", "C++", "Java", "JS", "C", "C#"]
index = np.arange(len(labels))
ratings = [5.168, 5.726, 14.988, 3.165, 11.857, 4.453]
plt.bar(index, ratings)
plt.xticks(index, labels)
plt.ylabel("Usage")
plt.title("Programming Language Usage") 
plt.show()
```

```python
labels = ["Python", "C++", "Java", "JS", "C", "C#"]
ratings = [5, 6, 15, 3, 12, 4]
 
plt.pie(ratings, labels=labels)
plt.title("Programming Language Usage") 
plt.axis("equal")
plt.show()
```

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
sinhs = np.sinh(x)
fig, ax = plt.subplots()
ax.plot(x, sinus, "r-o")
ax2 = ax.twinx()
ax2.plot(x, sinhs, "g--")
plt.show()
```
